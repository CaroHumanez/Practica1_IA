```python
from typing import List

def find_snake_on_grid(grid: List[str]) -> List[List[int]]:
    height = len(grid)
    width = len(grid[0])

    # Paso 1: encontrar la cabeza 'h'
    for y in range(height):
        for x in range(width):
            if grid[y][x] == 'h':
                current = (x, y)
                break
        else:
            continue
        break

    # Mapa de direcciones inversas: c칩mo se mueve la serpiente seg칰n el segmento actual
    direction_map = {
        '>': (-1, 0),  # se mueve hacia la izquierda (viene de la derecha)
        '<': (1, 0),   # se mueve hacia la derecha (viene de la izquierda)
        '^': (0, 1),   # se mueve hacia abajo (viene de arriba)
        'v': (0, -1),  # se mueve hacia arriba (viene de abajo)
    }

    path = [list(current)]

    while True:
        x, y = current
        found_next = False

        for dx, dy in direction_map.values():
            nx, ny = x + dx, y + dy
            if 0 <= nx < width and 0 <= ny < height:
                char = grid[ny][nx]
                if char in direction_map:
                    rev_dx, rev_dy = direction_map[char]
                    if (nx + rev_dx, ny + rev_dy) == (x, y):
                        current = (nx, ny)
                        path.append([nx, ny])
                        found_next = True
                        break

        if not found_next:
            break

    return path





[
  " >>h   ",
  " ^   v ",
  " ^<<<< ",
]
```

```python
from typing import List

# Funci칩n para encontrar el camino de la serpiente
def find_snake_on_grid(grid: List[str]) -> List[List[int]]:
    H, W = len(grid), len(grid[0])

    # 1. Buscar la cabeza 'h'
    for y in range(H):
        for x in range(W):
            if grid[y][x] == 'h':
                head = (x, y)
                break
        else:
            continue
        break

    path = [list(head)]
    x, y = head

    # 2. Mapa de flechas que indican la direcci칩n del siguiente segmento
    move_map = {
        '>': (1, 0),
        '<': (-1, 0),
        '^': (0, -1),
        'v': (0, 1)
    }

    # 3. Seguir la serpiente usando las flechas
    while True:
        found = False
        for dx, dy in move_map.values():
            nx, ny = x + dx, y + dy
            if 0 <= nx < W and 0 <= ny < H:
                c = grid[ny][nx]
                if c in move_map:
                    rev_dx, rev_dy = move_map[c]
                    if (nx + rev_dx, ny + rev_dy) == (x, y) and [nx, ny] not in path:
                        path.append([nx, ny])
                        x, y = nx, ny
                        found = True
                        break
        if not found:
            break

    return path


#Funci칩n para elegir la compa침칤a de taxi
def choosing_taxi(tf1, vqr1, tf2, vqr2):
    # Convertir strings a floats
    tf1 = float(tf1)
    vqr1 = float(vqr1)
    tf2 = float(tf2)
    vqr2 = float(vqr2)

    # Caso 1: funciones id칠nticas
    if tf1 == tf2 and vqr1 == vqr2:
        return "Either"

    # Caso 2: misma pendiente, comparar tarifas fijas
    if vqr1 == vqr2:
        if tf1 < tf2:
            return "Company 1"
        else:
            return "Company 2"

    # Caso 3: encontrar punto de cruce
    d = (tf2 - tf1) / (vqr1 - vqr2)

    # Si el cruce es en distancia negativa, no tiene sentido
    if d < 0:
        # Compara para cualquier valor positivo
        if tf1 + vqr1 * 1 < tf2 + vqr2 * 1:
            return "Company 1"
        else:
            return "Company 2"

    d = round(d, 10)  # precisi칩n

    if vqr1 < vqr2:
        return f"Company 2 when distance < {d}, Either when distance = {d}, Company 1 when distance > {d}"
    else:
        return f"Company 1 when distance < {d}, Either when distance = {d}, Company 2 when distance > {d}"


```
