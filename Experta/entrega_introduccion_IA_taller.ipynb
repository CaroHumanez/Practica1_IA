```python
from experta import KnowledgeEngine, Fact, Rule, NOT

# Definici칩n de los hechos
class Usuario(Fact):
    pass

class Genero_favorito(Fact):
    pass

class Idioma_preferido(Fact):
    pass

class Formato_preferido(Fact):
    pass

class Recomendacion(Fact):
    pass


```
